```python
import pandas as pd
```

```python
#csv into dataframe
df1 = pd.read_csv('daily_food_nutrition_dataset.csv')
df1
```

```python
#Remove duplicate food items from Food_Item column and reset index
df1 = df1.drop_duplicates(subset=["Food_Item"], keep="first")
df1 = df1.reset_index(drop=True)
#Remove not needed columns
df1 = df1.drop(columns=["Sodium (mg)", "Meal_Type", "Date", "User_ID", "Water_Intake (ml)", "Cholesterol (mg)"])
#Rename fats column to Fats (g)
df1 = df1.rename(columns={"Fat (g)": "Fats (g)", "Fiber (g)": "Fibre (g)"})
df1


```

```python
df2=pd.read_csv('output_with_sugars.csv')
df2 = df2.drop_duplicates(subset=["Food"], keep="first")
#rename column names and keep them as in df1
df2 = df2.rename(columns={"Food": "Food_Item", "Fat": "Fats (g)", "Calories": "Calories (kcal)", "Group": "Category", "Protein": "Protein (g)", "Carbs": "Carbohydrates (g)", "Fiber": "Fibre (g)"})
#Remove not needed columns
df2 = df2.drop(columns=["Measure", "Grams", "Sat.Fat"])
df2
```

```python
df4=pd.read_csv('cat.csv')
df4 = df4.drop_duplicates(subset=["Dish Name"], keep="first")
#Remove columns which are not in df1 and df2
df4 = df4.drop(columns=["Sodium (mg)","Calcium (mg)","Iron (mg)","Vitamin C (mg)","Folate (µg)"])
#Rename columns
df4 = df4.rename(columns={"Dish Name": "Food_Item", "Free Sugar (g)": "Sugars (g)"})
df4
```
